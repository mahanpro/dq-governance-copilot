```python
%sql
USE CATALOG dq_demo;
USE SCHEMA core;

CREATE TABLE IF NOT EXISTS dq_profiles (
  table_name   STRING,
  column_name  STRING,
  batch_date   DATE,
  row_count    LONG,
  null_count   LONG,
  null_fraction DOUBLE,
  distinct_count LONG,
  min_value    STRING,
  max_value    STRING,
  profile_ts   TIMESTAMP
)
USING delta;

CREATE TABLE IF NOT EXISTS dq_incidents (
  id                STRING,
  table_name        STRING,
  column_name       STRING,
  batch_date        DATE,
  ground_truth_id   STRING,   -- fk into dq_ground_truth.id, can be NULL
  incident_type     STRING,   -- e.g. 'NULL_SPIKE', 'DRIFT'
  severity          STRING,   -- e.g. 'LOW','MEDIUM','HIGH'
  details           STRING,
  detection_ts      TIMESTAMP -- for time-to-detection metric
)
USING delta;
```
